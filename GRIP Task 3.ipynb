```python
# Importing libraries in Python
import sklearn.datasets as datasets
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Loading the iris dataset
iris=datasets.load_iris()
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters. It also helps to form some simple classification models by drawing some simple lines or make linear separation in our data-set

```python
df.boxplot()
```

Boxplot shows which attributes has outliers.An outlier is a data point that is noticeably different from the rest.Outliers are basically those data point which statistically different from other so here sepalwidth feature has outliers

```python
q1,q2,q3 = np.quantile(df["sepal width (cm)"], 0.25), np.quantile(df["sepal width (cm)"], 0.50), np.quantile(df["sepal width (cm)"], 0.75)
IQR = q3 - q1
uw,lw = q3 + (1.5 * IQR),q1 - (1.5 * IQR)
```

IQR is Interquartile range so data points which are not in Interquartile range are outliers


```python
print(q1,q2,q3)
print(IQR)
print(uw,lw)
```

```python
df["sepal width (cm)"] = np.where(df["sepal width (cm)"] > uw, uw, df["sepal width (cm)"])
df["sepal width (cm)"] = np.where(df["sepal width (cm)"] < lw, lw, df["sepal width (cm)"])
```

```python
df.boxplot("sepal width (cm)")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train, y_test = train_test_split(df, y, test_size = 0.20, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

# Building ML-model:- Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
model = clf.fit(x_train, y_train)
```

```python
model
```

```python
y_pred = model.predict(x_test)
y_pred[:10]
```

```python
# Comparing Actual Vs Predicted Data

comp = pd.DataFrame({"Actual": y_test, "Predicted":y_pred})
comp
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print('Confusion Matrix is:\n', confusion_matrix(y_test, y_pred))
```

```python
print('Classification Report', classification_report(y_test, y_pred))
```
